IMPORTING LIBRARIES

```
import pandas as pd
import numpy as np
```

LOADING DATASETS

```
path = "//content//Movie_Id_Titles"
movie= pd.read_csv(path)
movie.head()
```

```
columnnames = ['user_id','item_id','ratings','timestamp']
path2 = "//content//u.data"
user = pd.read_csv(path2, sep='\t', names= columnnames)
user.head() 
```

Merging Datasets to single Dataframe

```
data = pd.merge(user ,movie ,  how = 'inner', on = 'item_id')
data
```

Pivot Table to check which user has given rating to which movie and how much rating he has given.

```
movie_pivot = data.pivot_table(index= 'user_id',columns='title', values='ratings')
movie_pivot 
```

Take input from the user to recommend movies accordingly


```
movie_name = input("MOVIE NAME FOR SIMILAR RECOMMENDATIONS : ")
data_movie = movie_pivot[movie_name]
print(data_movie.dropna())
```

Finding Correlation of all the movies with the movie entered by user

```
corr = movie_pivot.corrwith(data_movie)
similar_movies =pd.DataFrame(corr,columns=['correlation'])
similar_movies.sort_values('correlation', ascending = False ,inplace= True)
similar_movies.head()
```

```
df_ratings = pd.DataFrame(data.groupby('title')['ratings'].mean())
df_ratings
```

```
df_ratings['rate_count'] = pd.DataFrame(data.groupby('title')['ratings'].count())
df_ratings.sort_values(by = 'rate_count', ascending=False , inplace=True)
df_ratings
```

```
recommend= df_ratings.join(similar_movies)
recommend.sort_values('correlation', ascending=False , inplace=True)
recommend.head(10)
```

```
final = []
for i in range(len(recommend)): 
  if (recommend.iloc[i,1]> 15)== True:
   final.append(recommend.iloc[i])
```

```
final= pd.DataFrame(final)
final.head(10)  # Recommended movies
```
